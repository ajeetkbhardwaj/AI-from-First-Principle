KNNs
We store entire training dataset, when we needed to make prediction on new data point, we just locate k-most similar records to the stored records from training dataset. Then compute the similarity between new datapoints and recorded point using different metrices like euclidean distance etc. lead to identify new datapoints neighbors after that for making prediction we can take common outcome or just return the average of neighboring points.

Remarks
1. In KNNs, processing happens only during prediction time because during training we store the complete training dataset.
2. KNNs can be used to perform the both Regression as well as Classification tasks.
3. KNN refered as lazy learning methods by ml practioners

Objectives
1. Euclidean Distance
2. Get Neighbors
3. Making Predictions
4. KNNsClassification
5. KNNsRegression

```python
from matplotlib.colors import ListedColormap
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=1234
    )
```

```python
k = 3
clf = KNN(k=k)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
accuracy = KNN.accuracy(y_test, predictions)
print(f"KNN classification accuracy: {accuracy}")
```
